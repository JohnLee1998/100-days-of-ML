# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './Data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
pd.set_option('display.max_columns', None)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
dtype_select = ['int','float','float64']
# app_train['LIVINGAPARTMENTS_AVG'].dtypes #檢查到有Numpy的float64資料格式，跟python 的float不伊樣
# list(zip(app_train.columns ,list(app_train.dtypes.isin(dtype_select))))
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns: %s" % len(numeric_columns))
```

```python
# 檢視這些欄位的數值範圍#draw boxplot

fig, ax = plt.subplots(figsize=(20, 25), dpi = 150)
fig.subplots_adjust(hspace=1.2, wspace=1.2)
count = 0

for col in numeric_columns:
    count += 1
    fig.add_subplot((len(numeric_columns)//4)+1,4,count, title=col)
    df = pd.DataFrame(app_train[col])
    plot = df.boxplot(vert=False) 
    #有些ax.boxplot()會跑不出來是因為有nan,用pd.dropna()清過一遍之後，就可以用plt.boxplot()跑
    #但pd的argument有可以pass到matplotlib的 matplotlib.pyplot.boxplot()，所以上面可以用vert=False轉成橫的boxplot
    
plt.show()
```

### 從上面的圖檢查的結果，至少這三個欄位好像有點可疑: AMT_INCOME_TOTAL, REGION_POPULATION_RELATIVE, OBS_60_CNT_SOCIAL_CIRCLE
- AMT_INCOME_TOTAL 可能因為輸入錯誤，或是可能收入超出平常人非常多
- REGION_POPULATION_RELATIVE 代表居住地區人口很多，可能真的是人口非常密集或是輸入錯誤
- OBS_60_CNT_SOCIAL_CIRCLE 有多少人在申請者的社會環境中，在超過截止期限後30天倒帳。推測為輸入錯誤

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())
app_train['AMT_INCOME_TOTAL'].plot(kind='box')
plt.show()
# 繪製 Empirical Cumulative Density Plot (ECDF)
# 看ECDF時注意，y軸是指百分位數，所以當圖形的斜率很緩時，代表資料比數不多。相反的，線陡峭時代表那個值上的比數很多。
# ECDF 相較於直方圖可以避免binning bias（就是當bins取不同區間的count時，會有很誤導的效果）
"""
YOUR CODE HERE
"""
#先把資料做的值跟出現次數拉出來，接下來按照值來排列，最後看累積次數
cdf = app_train['AMT_INCOME_TOTAL'].value_counts().sort_index().cumsum()


plt.plot(list(cdf.index), cdf/cdf.max(), marker = 'o')
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 x 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()
```
